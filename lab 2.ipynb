```python
import pandas as pd
import numpy as np

# 1. Create and display a one-dimensional array-like object using Pandas module
data = pd.Series([1, 2, 3, 4, 5])
print("One-dimensional array-like object using Pandas:")
print(data)
print()

# 2. Convert a Pandas Series to Python list and its type
series = pd.Series([1, 2, 3, 4, 5])
list_from_series = series.tolist()
print("Pandas Series converted to Python list:", list_from_series)
print("Type of converted list:", type(list_from_series))
print()

# 3. Add, subtract, multiply, and divide two Pandas Series
series1 = pd.Series([2, 4, 6, 8, 10])
series2 = pd.Series([1, 3, 5, 7, 9])
addition = series1 + series2
subtraction = series1 - series2
multiplication = series1 * series2
division = series1 / series2
print("Addition:")
print(addition)
print("Subtraction:")
print(subtraction)
print("Multiplication:")
print(multiplication)
print("Division:")
print(division)
print()

# 4. Compare the elements of two Pandas Series
series3 = pd.Series([2, 4, 6, 8, 10])
series4 = pd.Series([1, 3, 5, 7, 10])
comparison = series3 == series4
print("Comparison of elements of two Pandas Series:")
print(comparison)
print()

# 5. Convert a dictionary to a Pandas series
dictionary = {'a': 100, 'b': 200, 'c': 300, 'd': 400, 'e': 800}
converted_series = pd.Series(dictionary)
print("Converted series:")
print(converted_series)
print()

# 6. Convert a NumPy array to a Pandas series
numpy_array = np.array([10, 20, 30, 40, 50])
converted_series = pd.Series(numpy_array)
print("Converted Pandas series:")
print(converted_series)
print()

# 7. Change the data type of given a column or a Series
series = pd.Series(['100', '200', 'python', '300.12', '400'])
converted_series = pd.to_numeric(series, errors='coerce')
print("Change the data type to numeric:")
print(converted_series)
print()

# 8. Convert the first column of a DataFrame to a Series
data = {'col1': [1, 2, 3, 4, 7, 11], 'col2': [4, 5, 6, 9, 5, 0], 'col3': [7, 5, 8, 12, 1, 11]}
df = pd.DataFrame(data)
series_from_df = df['col1']
print("1st column as a Series:")
print(series_from_df)
print(type(series_from_df))
print()

# 9. Convert a given Series to an array
series = pd.Series(['100', '200', 'python', '300.12', '400'])
array_from_series = series.values
print("Series to an array:")
print(array_from_series)
print(type(array_from_series))
print()

# 10. Create a DataFrame from a dictionary and display it
data = {'X': [78, 85, 96, 80, 86], 'Y': [84, 94, 89, 83, 86], 'Z': [86, 97, 96, 72, 83]}
df = pd.DataFrame(data)
print("DataFrame from a dictionary:")
print(df)
print()

# 11. Create and display a DataFrame from a specified dictionary data which has the index labels
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
             'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
             'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
             'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(exam_data, index=labels)
print("DataFrame with index labels:")
print(df)
print()

# 12. Display a summary of the basic information about a specified DataFrame and its data
print("Summary of the basic information about this DataFrame and its data:")
print(df.info())
print()

# 13. Get the first 3 rows of a given DataFrame
print("First three rows of the data frame:")
print(df.head(3))
print()

# 14. Select the 'name' and 'score' columns from the DataFrame
print("Select specific columns:")
print(df[['name', 'score']])
print()

# 15. Select specified columns and rows from a given data frame
print("Select specific columns and rows:")
print(df.loc[['b', 'd', 'f', 'g'], ['score', 'qualify']])
print()

# 16. Select the rows where the number of attempts in the examination is greater than 2
print("Number of attempts in the examination is greater than 2:")
print(df[df['attempts'] > 2])
print()

# 17. Count the number of rows and columns of a DataFrame
print("Number of Rows:", len(df))
print("Number of Columns:", len(df.columns))
print()

# 18. Select the rows where the score is missing (NaN)
print("Rows where score is missing:")
print(df[df['score'].isnull()])
print()

# 19. Select the rows where the score is between 15 and 20 (inclusive)
print("Rows where score is between 15 and 20 (inclusive):")
print(df[(df['score'] >= 15) & (df['score'] <= 20)])
print()
```
